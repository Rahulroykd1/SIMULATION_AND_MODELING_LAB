1.Basic Stats

```python
import numpy as np

# Exam scores
scores = np.array([93,83,69,62,75,93,73,77,65,65,78,90,94,78,57,76,56,78,84,92])

# Calculate mean
mean_score = np.mean(scores)

# Population variance (ddof=0)
population_variance = np.var(scores)

# Sample variance (ddof=1)
sample_variance = np.var(scores, ddof=1)

# Sample standard deviation
sample_std_dev = np.std(scores, ddof=1)

print("Basic Statistics:")
print(f"Mean: {mean_score:.2f}")
print(f"Population Variance: {population_variance:.2f}")
print(f"Sample Variance: {sample_variance:.2f}")
print(f"Sample Standard Deviation: {sample_std_dev:.2f}")
```
